```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import glob
```

```python
data_dir = '/Users/user/Dropbox (The University of Manchester)/metrics'

# model_name = 'vq_stokes_i'
model_name = 'vq_peak_polarisation'

metrics_dir = f'{data_dir}/{model_name}'

metrics = glob.glob(f'{metrics_dir}/*.npy')

# m = np.load(metrics[0])

data = []

for m in metrics:
    data.append({
        'epoch': int(m.split('/')[-1].split('_')[1].split('.')[0]),
        'value': np.load(m)
    })

# metrics = [
#     {'epoch': int(m.split('/')[-1].split('_')[1].split('.')[0]),
#      'metric': m.split('/')[-1].split('_')[0],
#      'value': pd.read_csv(m, header=None).values.flatten()[0]}
# ]
```

```python
df = pd.DataFrame(data)
df = df.sort_values(by='epoch')

plt.plot(df['epoch'], df['value'], marker='o', linestyle='-', markersize=1.)
plt.yscale('log')
plt.xlabel('Epoch')
plt.ylabel('Loss')
```
