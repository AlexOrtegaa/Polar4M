```python
import numpy as np
import matplotlib.pyplot as plt

def get_1d_sincos_pos_emb(seq_len, dim, temperature=10000):
    pos = np.arange(seq_len)[:, None]  # (seq_len, 1)
    i = np.arange(dim // 2)[None, :]   # (1, dim/2)
    omega = 1 / (temperature ** (i / (dim / 2)))

    pos_emb_sin = np.sin(pos * omega)  # (seq_len, dim/2)
    pos_emb_cos = np.cos(pos * omega)  # (seq_len, dim/2)

    pos_emb = np.concatenate([pos_emb_sin, pos_emb_cos], axis=1)  # (seq_len, dim)
    return pos_emb

seq_len = 100
dim = 64
pos_emb = get_1d_sincos_pos_emb(seq_len, dim)

plt.figure(figsize=(10, 6))
for i in range(0, dim, 8):
    plt.plot(pos_emb[:, i], label=f'Dim {i}')
plt.title('1D Sine-Cosine Positional Embeddings')
plt.xlabel('Position')
plt.ylabel('Embedding value')
plt.legend()
plt.show()
```
